# Package Coach

## Objective
This Python Coach is a jupyter script that challenges the user on his/her Python Pandas basic knowledge.

One question is loaded from excel file (source) and prints the category and a question. The user inputs a response.

Then the scripts shows the expected reply and after a self assessement increment in the source file a knowledge ratio per line/function.

Later on the questions will be oriented on the weaknesses of the user.

## How to use
Run All cells and enjoy!

Every 5 questions the script will ask if the user wants to stop.

```python
import pandas as pd
from IPython.display import Image, display
import random
import time
```

```python
class challenger():
    
    def __init__(self):
        print("To follow up with your progression you will be asked if your answer was correct,")
        print("input any character if you're right, but input empty if wrong")
        print("After a while I'll challenge your weaknesses :D")
        print(30* "-")
        
        self.emoji_neg = ['🤔', '🤨', '😐', '😑', '😶', '🙄', '😏', '😣', '😥', '😮', '🤐', '😯', 
                     '😪', '😫', '😴', '😌', '😒', '😓', '😔', '😕', '🙃', '😲', '☹️', '🙁', 
                     '😖', '😞', '😟', '😤', '😢', '😭', '😦', '😧', '😨', '😩', '🤯', '😬', 
                     '😰', '😱', '🥵', '🥶', '😳', '🤪', '😵', '😡', '😠', '🤬', '😷', '🤒', 
                     '🤕', '🤢', '🤮', '🤧', '🥴', '🥺', '🤥', '🤫', '🤭', '👿', '👹', '👺', 
                     '💀', '💩', '🙀', '😿', '😾', '👎', '🤞', '👆', '✋', '🖕', '💣', '🚽']
        
        self.emoji_pos = ['😀', '😁', '😂', '🤣', '😃', '😄', '😅', '😆', '😉', '😊', '😋', '😎', 
                     '😍', '😘', '🥰', '😗', '😙', '😚', '🙂', '🤗', '🤩', '😛', '😜', 
                     '😝', '🤤', '🤑', '😇', '🤠', '🤡', '🥳', '🧐', '🤓', '😈', '👻', '👽', 
                     '🤖', '😺', '😸', '😹', '😻', '😼', '😽', '🙌', '👏', '🤝', '👍', '👊', 
                     '✊', '🤛', '🤜', '✌️', '🤟', '🤘', '👌', '🖖', '🤙', '💪', '🙆', '🐍', 
                     '👵', '🦄', '🦖']
        
    def give_supportive_feedback(self, is_correct):
        if is_correct == '':
            print('Oops! ', self.emoji_neg[random.randrange(0,len(self.emoji_neg))])
        else:
            print('Nice! ', self.emoji_pos[random.randrange(0,len(self.emoji_pos))])
    
    
    def challenge_me(self):
        qcount = 0
        while True:
            qcount += 1
            # Ask if user wants to stop every 5 questions
            if qcount % 5 == 0:
                print(30* "#")
                stop = input(f"it's been {qcount} questions, do you want to stop ? [Y/N]")
                print("")
                if stop.lower() == 'y':  # stop if y
                    break
            
            # initiate
            df = pd.read_excel("Pandas_Cheat_Sheet.xlsx")
            
            # prepare a question (random line)
            line = random.randrange(0,df.shape[0])

            # build question
            category = df.loc[line, 'category']
            description = df.loc[line, 'description']
            image = df.loc[line, 'image']

            print(f"Category: {category}\nQuestion: '{description}'")
            if pd.notnull(image):  # add image if possible
                  display(Image(filename=f'images/{int(image)}.PNG'))

            # request an answer
            input('> ')

            # print expected answer
            answer = df.loc[line, 'answer']
            print(f"\nAnswer was:\n\t{answer}")

            # self assessment
            # input empty if wrong
            print('-- Self Assessment --')
            is_correct = input('\nPress enter if it was wrong, enter anything to show you joy elsewise!\n> ')
            if is_correct == '': 
                df.loc[line, 'total_wrong'] += 1
            
            self.give_supportive_feedback(is_correct) #print emoji
            
            # increment total asked
            df.loc[line, 'total_answer'] += 1
            df.loc[line, 'ratio_wrong'] = df.loc[line, 'total_wrong']/df.loc[line, 'total_answer']

            # end
            df.to_excel("Pandas_Cheat_Sheet.xlsx", index=False)
            del(df)
            print(30*"-","\n")
            time.sleep(0.75)
    
```

```python
q = challenger()
q.challenge_me()
```
